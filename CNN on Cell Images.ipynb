```python
# Unlike MNIST or CIFAR-10 datasets, Real life images are not going to be of the same shape and size
# Thus we need to find the average size and shape of the images dataset
```

```python
# Checking the dimension of the test images

dim1 = []
dim2 = []


for image_filename in os.listdir(test_path + "\\uninfected"):
    
#     "image_name" will give the name of every file in the uninfected images of test data set
# "test_path + "\\uninfected\\"" will give the path of the folder
# "imread" will read the files
    img = imread(test_path + "\\uninfected\\" + image_filename)
    
# "img.shape" will provide the shape of the picture
#     The shape of the image is in the form of the tuple (width,height,color)
# So we need to unpack it and save in the 3 variables
    d1,d2,colors = img.shape
    
# Appending the width to dim1 and height to dim2
    dim1.append(d1)
    dim2.append(d2)
```

```python
# We can see the width are different of all the images and so will be height

dim1
```

```python
# To visualize the image size

sns.jointplot(dim1,dim2)
```

```python
# We can see that majority of pictures lies between 100*120 to around 160*160
```

```python
# To check the average size of the picture

print(np.mean(dim1))
print(np.mean(dim2))
```

```python
# So we can see that average size of the picture is around 130*130
# and hence we can convert or resize the images to that value for training our model
# The shape will be (130*130*3)
# We'll get 130*130*30 = "50700" datapoints in our single file
```

```python
# To handle this much data we have to make batches of our images
# We also have to resize our images as discussed above
# "ImageDataGenerator" class will help us to do that

from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Checking the scale of the picture. In our case it already falls between 0-1 or else we,ll have to rescale it
# by dividing by 255 (by doing 1/255) in the instance of ImageDataGenerator

print(imread(para_cell).max())
print(imread(uninfected_cell).max())
```

```python
# To generate the instance of the ImageDataGenerator Class



# "rotation_range" rotates the images. Parameter given is in degrees
# the value of parameters has to be determined in accordance to our images and use case

# As our dataset size is smaller in size i.e. its less than 30000 images on whole
# It is much less than 60000 images and 50000 images in MNSIT and CIFAR-10 dataset respectively
# So, we need to randomize the images we have either by rotation or something
# And add those randomized images to our dataset as well to increase its size for the machine to learn effectively

# In our case scale of the picture already falls between 0-1 or else we,ll have to rescale it
# by dividing by 255 (by doing 1/255) in the instance of ImageDataGenerator

# "width_shift_range" and "height_shift_range" shifts the height and width of the image
# the parameter is in percentage form i.e. 0.1 stands for 10 percent and so on and so forth
# the value of parameters has to be determined in accordance to our images and use case

# "shear range" is cutting away the part of the image. Parameter given is in percentage

# "zoom_range" is for zooming the image. Parameter given is in percentage

# "horizontal_flip" and "vertical_flip" flips the picture. Parameter given is in True or False

# Fill mode is for filling the missing data after the transformation. we can leave it as blank with the padding of zeroes
# or we can also  fill it with the nearest data point/pixel values


image_gen = ImageDataGenerator(rotation_range = 20,
                              width_shift_range=0.1,
                              height_shift_range=0.1,
                              shear_range=0.1,
                              zoom_range=0.1,
                              horizontal_flip=True,
                              fill_mode="nearest")
```

### Generating many manipulated images from a directory


In order to use .flow_from_directory, you must organize the images in sub-directories. This is an absolute requirement, otherwise the method won't work. The directories should only contain images of one class, so one folder per class of images.

Structure Needed:

* Image Data Folder
    * Class 1
        * 0.jpg
        * 1.jpg
        * ...
    * Class 2
        * 0.jpg
        * 1.jpg
        * ...
    * ...
    * Class n

```python
# Number of images in training set
image_gen.flow_from_directory(train_path)
```

```python
# Number of images in test set
image_gen.flow_from_directory(test_path)
```
